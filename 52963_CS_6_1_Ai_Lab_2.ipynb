```python
# built-in function for list
list2 = [14, 59, 92, 73, 28]

# 1. Append
print("\n Question 1: append() Function")
list2.append(44)
print(list2)

# 2. Clear
print("\n Question 2: clear() Function")
list2.clear()
print(list2)

list2 = [14, 59, 92, 73, 28]

# 3. Copy
print("\n Question 3: copy() Function")
list2.copy()
print(list2)

# 4. Count
print("\n Question 4: count() Function")
list2.count(2)
print(list2)

# 5. Extend
print("\n Question 5: extend() Function")
list3 = [12,13,14]
list2.extend(list3)
print(list2)

# 6. index
print("\n Question 6: index() Function")
print("Before Changeing: ",list2[3])
list2 [3] = 55
print("After Changing: ",list2[3])

# 7. insert
print("\n Question 7: insert() Function")
list2.insert(11, 1)
print(list2)

list2.insert(1, 11)
print(list2)

# 8. pop
print("\n Question 8: pop() Function")
print(list2)

list2.pop()
print(list2)

# 9. remove
print("\n Question 9: remove() Function")
print("Original: ",list2)
list2.remove(12)
print("After Remove Element: ",list2)

# 10. Revese
print("\n Question 10: reverse() Function")
print("Original: ",list2)
print("Reversed: ",list2.reverse())

# 11. Revese
print("\n Question 11: sort() Function")
list2.sort()
print(list2)
```
