```python
import numpy as np

arr = np.arange(1, 13)  # sample array (1â€“12)

split_arrays = np.array_split(arr, 3)

total_sum = sum(a.sum() for a in split_arrays)

print("Original Array:", arr)
print("Split Arrays:", split_arrays)
print("Total Sum:", total_sum)
```

Question-3:

```python
# set1 = Even numbers 0â€“10
set1 = {x for x in range(11) if x % 2 == 0}

# set2 created using constructor
set2 = set(range(1, 11))

# remove 2nd and 5th element (index based after sorting for consistency)
sorted_set2 = sorted(set2)
set2.remove(sorted_set2[1])
set2.remove(sorted_set2[4])

# intersection
set3 = set1.intersection(set2)

print("Set1:", set1)
print("Set2 after removal:", set2)
print("Intersection (Set3):", set3)
```
