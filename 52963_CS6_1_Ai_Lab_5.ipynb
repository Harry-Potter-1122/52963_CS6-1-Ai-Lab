Question-1:

```python
def dfs(node, graph, visited, component):
    component.append(node)  # Store answer
    visited[node] = True  # Mark visited

    # Traverse to each adjacent node of a node
    for child in graph[node]:
        if not visited[child]:  # Check whether the node is visited or not
            dfs(child, graph, visited, component)  # Call the dfs recursively


if __name__ == "__main__":

    # Graph of nodes
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    node = 0
    visited = [False] * len(graph)
    component = []
    dfs(node, graph, visited, component)
    print(f"Following is the Depth-first search: {component}")
```

Question-2:

```python
from collections import deque

def bfs(start, graph, visited, component):
    queue = deque([start])
    visited[start] = True

    while queue:
        node = queue.popleft()
        component.append(node)  # Store answer

        # Traverse adjacent nodes
        for child in graph[node]:
            if not visited[child]:
                visited[child] = True
                queue.append(child)


if __name__ == "__main__":

    # Graph of nodes
    graph = {
        0: [1, 2, 3],
        1: [0],
        2: [0, 3, 4],
        3: [0, 2],
        4: [2]
    }

    node = 0  # Starting node
    visited = [False] * len(graph)
    component = []

    bfs(node, graph, visited, component)

    print(f"Following is the Breadth-first search: {component}")
```

Question-3:

```python
from collections import deque

# -------- DFS --------
def dfs(node, graph, visited, result):
    visited[node] = True
    result.append(node)

    for child in graph[node]:
        if not visited[child]:
            dfs(child, graph, visited, result)


# -------- BFS --------
def bfs(start, graph, visited, result):
    queue = deque([start])
    visited[start] = True

    while queue:
        node = queue.popleft()
        result.append(node)

        for child in graph[node]:
            if not visited[child]:
                visited[child] = True
                queue.append(child)


# -------- MAIN --------
if __name__ == "__main__":

    graph = {
        0: [2, 3],
        1: [2],
        2: [0, 1, 4],
        3: [0],
        4: [2]
    }

    start_node = 0

    # DFS
    visited_dfs = [False] * len(graph)
    dfs_result = []
    dfs(start_node, graph, visited_dfs, dfs_result)

    # BFS
    visited_bfs = [False] * len(graph)
    bfs_result = []
    bfs(start_node, graph, visited_bfs, bfs_result)

    print("DFS Traversal:", dfs_result)
    print("BFS Traversal:", bfs_result)
```
