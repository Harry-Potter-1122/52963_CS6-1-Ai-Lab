```python
# Loops in Python + table of input number

num = int(input("Enter a number: "))

for i in range(1,11):
  print(i, " x ", num, " = ", num*i)
```

```python
# print first 20 numbers

def prime_checker(n):
  if n == 1 or n == 0: return 0
  for i in range (2,n):
    if n % i == 0: return 0
  return 1

count = 1
for i in range (2,101):
  if prime_checker(i) == 1:
    print(count, ". ", i)
    count +=1
    if count == 21: break
```
