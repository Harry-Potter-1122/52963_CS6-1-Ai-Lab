```python
import heapq

goal = [[1,2,3],
        [4,5,6],
        [7,8,0]]

moves = [(0,1),(0,-1),(1,0),(-1,0)]

def heuristic(state):
    h = 0
    for i in range(3):
        for j in range(3):
            val = state[i][j]
            if val != 0:
                x = (val-1)//3
                y = (val-1)%3
                h += abs(x-i) + abs(y-j)
    return h

def find_zero(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i,j

def print_state(state):
    for row in state:
        print(row)
    print()

def solve(start):
    pq = []
    heapq.heappush(pq,(heuristic(start),0,start,[start]))
    visited = set()

    while pq:
        f,g,state,path = heapq.heappop(pq)

        if state == goal:
            print("Goal Reached!\n")
            print("Steps to reach goal:\n")
            for step in path:
                print_state(step)
            return

        visited.add(str(state))
        x,y = find_zero(state)

        for dx,dy in moves:
            nx,ny = x+dx,y+dy

            if 0<=nx<3 and 0<=ny<3:
                new_state = [row[:] for row in state]
                new_state[x][y],new_state[nx][ny] = new_state[nx][ny],new_state[x][y]

                if str(new_state) not in visited:
                    heapq.heappush(pq,(g+1+heuristic(new_state),
                                       g+1,
                                       new_state,
                                       path+[new_state]))

start = [[1,2,3],
         [4,0,6],
         [7,5,8]]

solve(start)
```

```python
import heapq

graph = {
    'A':{'B':1,'C':3},
    'B':{'D':3,'E':6},
    'C':{'E':2},
    'D':{'F':4},
    'E':{'F':1},
    'F':{}
}

heuristic = {
    'A':7,
    'B':6,
    'C':2,
    'D':1,
    'E':1,
    'F':0
}

def astar(start,goal):
    pq = []
    heapq.heappush(pq,(heuristic[start],0,start,[start]))
    visited = set()

    while pq:
        f,g,node,path = heapq.heappop(pq)

        if node == goal:
            print("Shortest Path:",path)
            print("Cost:",g)
            return

        visited.add(node)

        for neighbor,cost in graph[node].items():
            if neighbor not in visited:
                heapq.heappush(pq,(g+cost+heuristic[neighbor],
                                   g+cost,
                                   neighbor,
                                   path+[neighbor]))

astar('A','F')
```
